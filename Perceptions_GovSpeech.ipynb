#Introduction






The purpose of this code is to gain insight into governmental perceptions towards migrants in Latin-American countries.

Analysis was conducted to generate word clouds and sentiment scores.

Namely, this code looks at Federal Government speeches from English-speaking Latin-American Countries (LAC).

This code scrapes governmental speeches from the following nations:
* Bahamas
* Barbados
* Belize
* Guyana
* Jamaica
* Trinidad y Tobago


#Methodology

Title of speeches, Date published, and the Speech itself were scraped from each speech.

A total of 4242 speeches were collected, which were narrowed down to _____________ speeches containing key words relating to migration.

The key words are: "Diaspora", "Migration", "Immigrant", "Immigration", "Migrant", "Expatriate", "Expat", "Asylum", "Refugee", and "Returnee".

Word clouds were first generated for each source to gain an initial visual understanding of the contents spoken and to extract any additional stopwords required to append to NLTK's list of stopwords.

This was then followed by sentiment analysis on the article level. 2 methods were used to extract segments of text for analysis.
1. Only sentences containing the keywords were extracted.
2. Sentences related to the key words were extracted using the [Zero-shot-classification](https://huggingface.co/tasks/zero-shot-classification) from Hugging Face.

Sentiments were generated using [lxyuan/distilbert-base-multilingual-cased-sentiments-student](https://huggingface.co/lxyuan/distilbert-base-multilingual-cased-sentiments-student?text=quiero+chocolate) for each sentence and averaged out to gain the overall sentiment for each document.

These sentiments were plotted out for each source to generate a time-series graph for sources of which we were able to scrape published dates from. Otherwise, bar charts were instead generated.

```python
def DriverSet():
    """Open Driver"""
    options = webdriver.ChromeOptions()
    options.add_argument("--headless")
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument("--no-sandbox")

    driver = webdriver.Chrome(
      options=options
    )
    return driver
```

```python
def get_link(by, path):
    """Find link in the page with given text"""
    element = driver.find_element(by,path)
    return element.get_attribute("href")

def get_link_list(by, name):
    """Get list of links in all element by name"""
    element_list = []
    try:
        all_elements = driver.find_elements(by,name)
        element_list = [x.get_attribute('href') for x in all_elements
                        if len(x.get_attribute('href')) > 0]
    except (NoSuchElementException, WebDriverException) as e:
        print(e)
    return element_list

def get_list(by, name):
    """Get list of text in all element by name"""
    element_list = []
    try:
        all_elements = driver.find_elements(by,name)
        element_list = [x.text for x in all_elements if len(x.text) > 0]
    except (NoSuchElementException, WebDriverException) as e:
        print(e)
    return element_list

def get_onclick_list(by, name, root):
    """Get list of text in all element with attribute onclick"""
    element_list = []
    try:
        all_elements = driver.find_elements(by,name)
        element_list = [root+x.get_attribute('onclick').split("'")[1]
                        for x in all_elements if len(x.text) > 0]
    except (NoSuchElementException, WebDriverException) as e:
        print(e)
    return element_list
```

##Functions for Processing Data

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence).encode('utf-8'), deacc=True))  # deacc=True removes punctuations

def flatten(xs):
    for x in xs:
        if isinstance(x, Iterable) and not isinstance(x, (str, bytes)):
            yield from flatten(x)
        else:
            yield x

# Define functions for stopwords, bigrams, trigrams and lemmatization
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent))
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```
